```python
a = np.array([[1,2], [3,4], [5,6], [7,8]])
b = a[:, np.newaxis, :] - a[np.newaxis, :, :]
print(b)
c = np.sum(b**2, axis=-1)
print(c)
overlap = c <=8
np.fill_diagonal(overlap, False)
print(overlap)
```
